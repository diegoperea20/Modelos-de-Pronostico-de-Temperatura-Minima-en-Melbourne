```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Concatenate, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

# Cargar y preparar los datos
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv'
df = pd.read_csv(url, parse_dates=['Date'], index_col='Date')

# Crear características de tiempo
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year

# Normalizar los datos
scaler_temp = MinMaxScaler()
scaler_features = MinMaxScaler()

temp_scaled = scaler_temp.fit_transform(df[['Temp']])
features_scaled = scaler_features.fit_transform(df[['day', 'month', 'year']])

# Crear secuencias de datos
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 10
X_temp, y = create_sequences(temp_scaled, seq_length)

# Dividir los datos en conjuntos de entrenamiento y prueba
train_size = int(len(X_temp) * 0.8)
X_train_temp, X_test_temp = X_temp[:train_size], X_temp[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

X_train_features = features_scaled[seq_length:train_size+seq_length]
X_test_features = features_scaled[train_size+seq_length:]

# Crear el modelo
temp_input = Input(shape=(seq_length, 1))
features_input = Input(shape=(3,))

lstm = LSTM(64)(temp_input)
dense1 = Dense(32, activation='relu')(lstm)

concat = Concatenate()([dense1, features_input])
dense2 = Dense(16, activation='relu')(concat)
output = Dense(1)(dense2)

model = Model(inputs=[temp_input, features_input], outputs=output)

# Compilar el modelo
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

# Entrenar el modelo
history = model.fit(
    [X_train_temp, X_train_features], y_train,
    validation_data=([X_test_temp, X_test_features], y_test),
    epochs=100, batch_size=32, verbose=1
)

# Guardar el modelo
model.save('model.h5')

# Función para predecir la temperatura para una fecha dada
def predict_temperature(date_str):
    date = datetime.strptime(date_str, '%Y-%m-%d')
    input_features = np.array([[date.day, date.month, date.year]])
    input_features_scaled = scaler_features.transform(input_features)

    # Obtener los últimos 10 días de temperatura
    last_10_days = temp_scaled[-10:].reshape(1, 10, 1)

    prediction_scaled = model.predict([last_10_days, input_features_scaled])
    prediction = scaler_temp.inverse_transform(prediction_scaled)[0][0]
    return prediction

# Ejemplo de uso
print(f"Temperatura predicha para 1991-01-01: {predict_temperature('1991-01-01'):.2f}")
print(f"Temperatura predicha para 1998-01-01: {predict_temperature('1998-01-01'):.2f}")

# Visualizar el rendimiento del modelo
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='validation')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()

# Visualizar predicciones vs valores reales
dates_to_predict = pd.date_range(start='1991-01-01', end='1991-12-31', freq='D')
predictions = [predict_temperature(date.strftime('%Y-%m-%d')) for date in dates_to_predict]

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Temp'], label='Datos reales', alpha=0.7)
plt.plot(dates_to_predict, predictions, label='Predicciones', color='red')
plt.title('Temperaturas reales vs Predicciones')
plt.xlabel('Fecha')
plt.ylabel('Temperatura')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

def evaluate_model(model, X_test_temp, X_test_features, y_test, scaler_temp):
    # Hacer predicciones
    y_pred_scaled = model.predict([X_test_temp, X_test_features])

    # Invertir la normalización
    y_pred = scaler_temp.inverse_transform(y_pred_scaled)
    y_true = scaler_temp.inverse_transform(y_test.reshape(-1, 1))

    # Calcular métricas
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    r2 = r2_score(y_true, y_pred)

    print(f"MSE: {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"MAE: {mae:.4f}")
    print(f"MAPE: {mape:.4f}%")
    print(f"R²: {r2:.4f}")

    # Graficar predicciones vs valores reales
    plt.figure(figsize=(12, 6))
    plt.plot(y_true, label='Valores reales')
    plt.plot(y_pred, label='Predicciones')
    plt.title('Predicciones vs Valores reales')
    plt.xlabel('Tiempo')
    plt.ylabel('Temperatura')
    plt.legend()
    plt.show()

    # Graficar residuos
    residuals = y_true - y_pred
    plt.figure(figsize=(12, 6))
    plt.plot(residuals)
    plt.title('Residuos')
    plt.xlabel('Tiempo')
    plt.ylabel('Error')
    plt.axhline(y=0, color='r', linestyle='--')
    plt.show()

    # Histograma de residuos
    plt.figure(figsize=(12, 6))
    plt.hist(residuals, bins=50)
    plt.title('Distribución de residuos')
    plt.xlabel('Error')
    plt.ylabel('Frecuencia')
    plt.show()

# Después de entrenar el modelo
evaluate_model(model, X_test_temp, X_test_features, y_test, scaler_temp)
```
